Task 1

```python
material_cost = float(input("Enter the cost of the material per metre: "))
metres = float(input("Enter the number of metres required: "))

cloth_total = material_cost * metres
making_cost = 6.50 * metres
installation_cost = 45
total_cost = cloth_total + making_cost + installation_cost

print("The total cost of the cloth is $", cloth_total)
print("The cost of making the cloth into curtains is $", making_cost)
print("The total cost including installation is $", total_cost)
```

Task 2

```python
distance = float(input("Enter the distance travelled (km): "))
hours = float(input("Enter the number of hours worked: "))
parts_cost = float(input("Enter the cost of the parts: "))

travel_cost = distance * 1.50
labour_cost = hours * 60
total_cost = travel_cost + labour_cost + parts_cost

print("The cost of the travel is $", travel_cost)
print("The cost of the labour is $", labour_cost)
print("The total cost of the bill is $", total_cost)
```

Task 3

```python
PRICE_PER_TONNE = 65
TRAILER_COST = 25

topsoil_tonnes = float(input("Enter the amount of topsoil purchased (in tonnes): "))

cost_of_topsoil = topsoil_tonnes * PRICE_PER_TONNE
total_cost = cost_of_topsoil + TRAILER_COST

print("BILL")
print(f"Amount of topsoil purchased: {topsoil_tonnes} tonnes")
print(f"Cost of topsoil (@ $65 per tonne): ${cost_of_topsoil:.2f}")
print(f"Trailer cost: ${TRAILER_COST:.2f}")
print(f"TOTAL COST: ${total_cost:.2f}")
```

Task 4


```python
price = float(input("Enter the full price of the purchases: "))

discount = price * 0.15
final_price = price - discount

print(" PURCHASE SUMMARY")
print(f"Initial price: ${price:.2f}")
print(f"Discount (15%): ${discount:.2f}")
print(f"Final price: ${final_price:.2f}")
```

Task5

```python
CHAIR_PRICE = 89
DESK_PRICE = 148

chairs = int(input("Enter the number of chairs purchased: "))
desks = int(input("Enter the number of desks purchased: "))

total_chair_cost = chairs * CHAIR_PRICE
total_desk_cost = desks * DESK_PRICE
total_cost = total_chair_cost + total_desk_cost

print("\n----- PURCHASE SUMMARY -----")
print(f"Chairs purchased: {chairs}  |  Cost: ${total_chair_cost:.2f}")
print(f"Desks purchased:  {desks}  |  Cost: ${total_desk_cost:.2f}")
print(f"TOTAL COST: ${total_cost:.2f}")
```
