Task 1

```python
videos = int(input("Enter number of videos rented: "))

if videos <= 10:
    cost = videos * 2
else:
    cost = 10 + (videos - 10) * 2

print("Total cost: $",cost)
```

Task 4

```python
units = float(input("Enter number of gas units used: "))


rate_first = 0.0745
rate_extra = 0.0685

if units <= 55:
    cost = units * rate_first
else:
    cost = 55 * rate_first + (units - 55) * rate_extra

print("Your monthly gas bill is: $", round(cost, 2))
```
